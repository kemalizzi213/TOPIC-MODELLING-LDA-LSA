# Cleaning

```python
import string
import re #regex library
```

```python
# ------ Cleaning ---------

nltk.download('punkt')

def remove_tweet_special(text):
    # remove tab, new line, ans back slice
    text = text.replace('\\t'," ").replace('\\n'," ").replace('\\u'," ").replace('\\',"")
    # remove non ASCII (emoticon, chinese word, .etc)
    text = text.encode('ascii', 'replace').decode('ascii')
    # remove mention, link, hashtag
    text = ' '.join(re.sub("([@#][A-Za-z0-9]+)|(\w+:\/\/\S+)"," ", text).split())
    # remove incomplete URL
    return text.replace("http://", " ").replace("https://", " ")

uinjkt_clean['abstrak'] = uinjkt_clean['abstrak'].apply(remove_tweet_special)

#remove number
def remove_number(text):
    return  re.sub(r"\d+", "", text)

uinjkt_clean['abstrak'] = uinjkt_clean['abstrak'].apply(remove_number)

#remove punctuation
def remove_punctuation(text):
    return text.translate(str.maketrans("","",string.punctuation))

uinjkt_clean['abstrak'] = uinjkt_clean['abstrak'].apply(remove_punctuation)

#remove whitespace leading & trailing
def remove_whitespace_LT(text):
    return text.strip()

uinjkt_clean['abstrak'] = uinjkt_clean['abstrak'].apply(remove_whitespace_LT)

#remove multiple whitespace into single whitespace
def remove_whitespace_multiple(text):
    return re.sub('\s+',' ',text)

uinjkt_clean['abstrak'] = uinjkt_clean['abstrak'].apply(remove_whitespace_multiple)

# remove single char
def remove_singl_char(text):
    return re.sub(r"\b[a-zA-Z]\b", "", text)

uinjkt_clean['abstrak'] = uinjkt_clean['abstrak'].apply(remove_singl_char)
```

```python
from sklearn.metrics import pairwise_distances
from sklearn.metrics import silhouette_score
from sklearn.metrics import classification_report, accuracy_score
```

```python
def topic_coherence(model, vectors):
    topic_coherence_scores = []
    for i in range(model.components_.shape[0]):
        topic_terms = [vocab[idx] for idx in model.components_[i].argsort()[-6:][::-1] if vocab[idx].isalnum()]
        topic_vectors = vectors[:, i]  # Mengubah baris ini
        pairwise_dists = pairwise_distances(topic_vectors[:, np.newaxis], metric='cosine')  # Mengubah baris ini
        avg_coherence = 0
        for j in range(len(topic_terms)):
            coherence_score = (pairwise_dists.sum() - pairwise_dists[j, :].sum()) / (len(topic_terms) - 1)
            avg_coherence += coherence_score
        avg_coherence /= len(topic_terms)
        topic_coherence_scores.append(avg_coherence)
    return topic_coherence_scores
```

```python
topic_coherence_scores = topic_coherence(lsa, lsa_matrix)
print("Topic Coherence Scores:")
for i, score in enumerate(topic_coherence_scores):
    print("Topic", i+1, ":", score)
```

```python
#  Mengukur Topic Significance Metrics
def topic_significance(model):
    topic_significance_scores = []
    total_variance = np.sum(model.explained_variance_)
    for i in range(model.components_.shape[0]):
        topic_variance = model.explained_variance_[i]
        topic_significance = topic_variance / total_variance
        topic_significance_scores.append(topic_significance)
    return topic_significance_scores

topic_significance_scores = topic_significance(lsa)
print("Topic Significance Scores:")
for i, score in enumerate(topic_significance_scores):
    print("Topic", i+1, ":", score)
```

```python
# Mengukur Topic Diversity Metrics
def topic_diversity(model):
    topic_diversity_scores = []
    for i in range(model.components_.shape[0]):
        topic_vector = model.components_[i]
        topic_diversity = np.sqrt(np.sum(np.square(topic_vector)))
        topic_diversity_scores.append(topic_diversity)
    return topic_diversity_scores

topic_diversity_scores = topic_diversity(lsa)
print("Topic Diversity Scores:")
for i, score in enumerate(topic_diversity_scores):
    print("Topic", i+1, ":", score)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
# import pandas as pd
# from sklearn.decomposition import LatentDirichletAllocation
from gensim.models import CoherenceModel
from sklearn.metrics import classification_report
# import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Topic coherence metric
def compute_coherence(lda_model, documents, vocab):
    vectorizer = CountVectorizer(vocabulary=vocab)
    doc_term_matrix = vectorizer.transform(documents)
    topic_term_matrix = lda_model.components_

    # Hitung skor UMass coherence
    similarity_matrix = cosine_similarity(doc_term_matrix, topic_term_matrix)
    coherence_score = np.mean(np.triu(similarity_matrix, k=1))

    return coherence_score
```

```python
coherence_score = compute_coherence(lda, documents, feature_names)
print("Topic Coherence Score (UMass):", coherence_score)
```

```python
def compute_topic_coherence(lda_model, documents, feature_names):
    vectorizer = CountVectorizer(vocabulary=feature_names)
    doc_term_matrix = vectorizer.transform(documents)
    topic_term_matrix = lda_model.components_

    num_topics = lda_model.n_components
    topic_coherences = np.zeros(num_topics)
    for topic in range(num_topics):
        topic_terms = topic_term_matrix[topic].reshape(1, -1)
        similarity_matrix = cosine_similarity(doc_term_matrix, topic_terms)
        topic_coherences[topic] = np.mean(similarity_matrix)

    return topic_coherences

coherence_scores = compute_topic_coherence(lda, documents, feature_names)
for topic, coherence in enumerate(coherence_scores):
    print("Topic", topic+1, "Coherence Score (UMass):", coherence)
```

```python
# Topic significance metric
def compute_topic_significance(lda_matrix):
    topic_significance = lda_matrix.sum(axis=0) / lda_matrix.sum()
    return topic_significance

topic_significance = compute_topic_significance(lda_matrix)
print("Topic Significance:", topic_significance)
```

```python
print("Topic Significance Scores:")
for i, score in enumerate(topic_significance):
    print("Topic", i+1, ":", score)
```

```python
# import numpy as np
from scipy.spatial import distance

def topic_diversity(model):
    num_topics = model.components_.shape[0]
    topic_diversity_scores = np.zeros(num_topics)

    for i in range(num_topics):
        topic_vector = model.components_[i]
        jsd_scores = []
        for j in range(num_topics):
            if i != j:
                jsd = 0.5 * (distance.jensenshannon(topic_vector, model.components_[j]) +
                            distance.jensenshannon(model.components_[j], topic_vector))
                jsd_scores.append(jsd)
        topic_diversity_scores[i] = np.mean(jsd_scores)

    return topic_diversity_scores

topic_diversity_scores = topic_diversity(lda)
print("Topic Diversity Scores:")
for i, score in enumerate(topic_diversity_scores):
    print("Topic", i+1, ":", score)
```
